#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
wiki_lang = requests.get(url)
wiki_languages_soup = BeautifulSoup(wiki_lang.content, "html.parser")
wiki_languages = wiki_languages_soup.find_all("table",class_="wikitable sortable")[0]
wiki_languages_rows = wiki_languages.find_all('tr')
wiki_languages_rowstabulated = [i.text.strip().split("\n") for i in wiki_languages_rows]
wiki_languages_columns = wiki_languages_rowstabulated[0:1]
wiki_languages_rowsrows = wiki_languages_rowstabulated[1:]
wiki_languages_columns = [j for i in wiki_languages_columns for j in i if i!=""]
df_wiki_languages = pd.DataFrame(wiki_languages_rowsrows, columns= wiki_languages_columns)
df_wiki_languages
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
imdb = requests.get(url)
imdb_soup = BeautifulSoup(imdb.content, "html.parser")
imdb_250 = imdb_soup.find_all("tr")
imdb_250_rows = [i.get_text() for i in imdb_250[1:]]
imdb_250_rows = [i.strip().split("\n") for i in imdb_250_rows]
imdb_250_rows_df = pd.DataFrame(imdb_250_rows)
imdb_250_rows_df = (imdb_250_rows_df.drop(imdb_250_rows_df.columns[6:19], axis = 1)).drop([0,3,4], axis = 1)
imdb_250_rows_df.columns = ["Title","Date release","Rating"]
imdb_250_directors = [i['title'].split(" (") for i in imdb_soup.find_all('a', title=True)]
imdb_250_directors = [i[0] for i in imdb_250_directors]
imdb_250_directors = imdb_250_directors[:-5]
imdb_250_rows_df.insert(loc=2, column="Director", value=imdb_250_directors)
imdb_250_rows_df
````

####  Display the 100 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.
***Hint:*** Here the displayed number of earthquakes per page is 20, but you can easily move to the next page by looping through the desired number of pages and adding it to the end of the url.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/?view='

# This is how you will loop through each page:
number_of_pages = int(100/50)
each_page_urls = []

for n in range(1, number_of_pages+1):
    link = url+str(n)
    each_page_urls.append(link)
    
each_page_urls
### It was showing 50 earthquakes per page. I had to change number_of_pages.
````

````python
# your code here
eq_df = pd.DataFrame()
eq_dt = pd.DataFrame()
for i in each_page_urls:
    eq = requests.get(i)
    eq_soup = BeautifulSoup(eq.content ,"html.parser")
    eq_info = eq_soup.find_all("td")
    eq_text = [i.get_text() for i in eq_info[21:]]
    eq_text = [i.replace("\xa0","") for i in eq_text]
    eq_text = [eq_text[i:i+13] for i in range(0,len(eq_text),13)]
    eq_text = eq_text[:-1]
    eq_date_time = [i[12].split(" ") for i in eq_text]
    eq_data = [i[0:12] for i in eq_text]
    eq_df = eq_df.append(eq_data, ignore_index = True)
    eq_dt = eq_dt.append(eq_date_time, ignore_index = True)
eq_df.drop([0,1,2,3,9], axis=1, inplace = True)
eq_df.rename(columns={4:"Lat. coord.",5:"Lat. orient.", 6:"Long. coord.",7:"Long. orient.",8:"Depth(km)",10:"Magnitude", 11: "Region"}, inplace=True)
eq_df["Date"],eq_df["Time"] = [eq_dt[0], eq_dt[1]]
eq_df
````
